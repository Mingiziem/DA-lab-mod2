#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(12,4))
ax1.plot(fitbit.groupby('Months_encoded').mean()['Steps'])
ax1.set_xlabel('Month')     
ax1.set_ylabel('Avg. steps')
ax1.set_xticks(fitbit['Months_encoded'].unique())
ax1.set_title('Avg. steps by month')


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
ax2.bar(fitbit.groupby('Work_or_Weekend').sum().index, fitbit.groupby('Work_or_Weekend').mean()['Steps'])
ax2.set_xlabel('Weekend or Workday')     
ax2.set_ylabel('Avg. steps')
ax2.set_title('Avg. steps during workdays and weekends')
ax2.set_xticks([0,1])
ax2.set_xticklabels(['Work', 'Weekend'])
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
fig, axes = plt.subplots(3, 1,figsize=(6,10))

for col, axe in zip(['Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active'], axes):
    axe.scatter(fitbit.groupby(col).mean().index, fitbit.groupby(col).mean()['Steps'])
    axe.grid(axis='both')
    axe.set_xlabel(col)
    axe.set_ylabel('Steps')
```
